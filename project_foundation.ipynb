```python
df=df.drop(columns=['Date_1'])
```

```python
df['Day']=df['Date'].dt.day
```

```python
df.head(10)
```

```python
X=df.drop(['Close','Date'],axis=1)
Y=df['Close']
featues_df=X.columns
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(random_state=10)
model.fit(X,Y)
```

```python
from sklearn.feature_selection import RFE
rfe_model=RFE(model,n_features_to_select=7)
fit=rfe_model.fit(X,Y)

print(fit.support_)
print(fit.ranking_)


SELECTED_FEATUES=[featues_df[i] for i in range(len(featues_df)) if fit.support_[i]]

featues_df_names=pd.DataFrame({'FEATURES NAMES':SELECTED_FEATUES})

featues_df_names

```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.40,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model_lG=LinearRegression()
model_lG.fit(X_train,Y_train)
```

```python
pred_train=model_lG.predict(X_train)
pred_test=model_lG.predict(X_test)
```

```python
plt.scatter(Y_train,pred_train,color='blue')
plt.plot(Y_train,Y_train,color='red')
```

```python
plt.scatter(Y_test,pred_test,color='blue')
plt.plot(Y_train,Y_train,color='red')
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
R2_SCORE=r2_score(Y_test,pred_test)
print(f'THE R2 SCORE OF LINEAR MODEL IS {R2_SCORE}')
MSE=mean_squared_error(Y_test,pred_test)

print(f'THE MEAN SQUARED ERROR OF LINEAR MODEL IS {MSE}')

MAE=mean_absolute_error(Y_test,pred_test)
print(f'THE MEAN ABSOLUTE ERROR {MAE}')



```

```python
from sklearn.tree import DecisionTreeRegressor
model_dt=DecisionTreeRegressor(random_state=10)
model_dt.fit(X_train,Y_train)
```

```python
pred_train_dt=model_dt.predict(X_train)
pred_test_dt=model_dt.predict(X_test)
```

```python
plt.scatter(Y_train,pred_train_dt,color='green')
plt.plot(Y_train,Y_train,color='red')
```

```python
plt.scatter(Y_test,pred_test_dt,color='green')
plt.plot(Y_train,Y_train,color='red')
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

R2_SCORE=r2_score(Y_test,pred_test_dt)
print(f'THE R2 SCORE OF DECISION TREE MODEL IS {R2_SCORE}')

MSE=mean_squared_error(Y_test,pred_test_dt)

print(f'THE MEAN SQUARED ERROR OF DECISION TREE MODEL IS{MSE}')

MAE=mean_absolute_error(Y_test,pred_test_dt)
print(f'THE MEAN ABSOLUTE ERROR OF DECISION TREE MODEL {MAE}')



```
